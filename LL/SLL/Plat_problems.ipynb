```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Linkedlist:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node


def merge_sorted(list1, list2):
    dummy = Node(0)
    tail = dummy

    while list1 and list2:
        if list1.data <= list2.data:
            tail.next = list1
            list1 = list1.next
        else:
            tail.next = list2
            list2 = list2.next
        tail = tail.next

    if list1:
        tail.next = list1
    if list2:
        tail.next = list2

    return dummy.next



array1 = list(map(int, input().split()))
array2 = list(map(int, input().split()))

ll1 = Linkedlist()
ll2 = Linkedlist()

for i in array1:
    if i == -1:
        break
    else:
        ll1.append(i)

for i in array2:
    if i == -1: 
        break
    else:
        ll2.append(i)

merged_head = merge_sorted(ll1.head, ll2.head)


temp = merged_head
while temp:
    print(temp.data, end=" ")
    temp = temp.next
```
