## ‚úÖ 1. What is a Singly Linked List?

A **Singly Linked List (SLL)** is a **linear data structure** in which elements (nodes) are connected using **pointers**.
Each node stores:

* ‚úÖ **Data**
* ‚úÖ **Address of the next node**

Traversal is possible **only in one direction**.

---

## ‚úÖ 2. Structure of a Node in SLL

Each node consists of **two parts**:

| Part | Description                                     |
| ---- | ----------------------------------------------- |
| Data | Stores the actual value                         |
| Next | Stores the reference (address) of the next node |

```
[ Data | Next ] ‚Üí [ Data | Next ] ‚Üí [ Data | None ]
```

---

## ‚úÖ 3. Main Components of SLL

| Component | Meaning                      |
| --------- | ---------------------------- |
| Head      | Points to the first node     |
| Tail      | Points to the last node      |
| Length    | Stores total number of nodes |

---

## ‚úÖ 4. Basic Properties of SLL

* ‚úÖ Dynamic size (can grow and shrink)
* ‚úÖ Memory is **not stored contiguously**
* ‚úÖ Nodes are connected using **links**
* ‚úÖ Last node points to **None**
* ‚úÖ Traversal happens **only forward**

---

## ‚úÖ 5. Why Do We Use Singly Linked Lists?

* ‚úÖ Efficient **insertion & deletion**
* ‚úÖ Saves memory compared to doubly linked list
* ‚úÖ Useful when **dynamic memory** is required
* ‚úÖ Avoids shifting elements like in arrays

---

## ‚úÖ 6. Basic Operations on SLL

| Operation          | Description                |
| ------------------ | -------------------------- |
| Append             | Insert at end              |
| Prepend            | Insert at beginning        |
| Insert at Position | Insert at a specific index |
| Delete at End      | Remove last node           |
| Delete at Start    | Remove first node          |
| Traversal          | Visit all nodes            |
| Searching          | Find an element            |
| Length             | Count number of nodes      |

---

## ‚úÖ 7. Time Complexity of SLL Operations

| Operation         | Time |
| ----------------- | ---- |
| Insertion at Head | O(1) |
| Insertion at Tail | O(1) |
| Deletion at Head  | O(1) |
| Deletion at Tail  | O(n) |
| Searching         | O(n) |
| Traversal         | O(n) |

---

## ‚úÖ 8. Advantages of Singly Linked List

‚úÖ Dynamic memory allocation
‚úÖ Easy insertion and deletion
‚úÖ Saves memory (only one pointer used)
‚úÖ No memory wastage like dynamic arrays

---

## ‚úÖ 9. Disadvantages of Singly Linked List

‚ùå Cannot move backward <br>
‚ùå Extra memory for pointer <br>
‚ùå Accessing elements is slower than arrays <br>
‚ùå Binary search not possible directly <br>

## ‚úÖ 12. Difference Between Array and SLL

| Feature   | Array      | SLL            |
| --------- | ---------- | -------------- |
| Memory    | Continuous | Non-Continuous |
| Size      | Fixed      | Dynamic        |
| Insertion | Slow       | Fast           |
| Deletion  | Slow       | Fast           |
| Access    | Random     | Sequential     |
