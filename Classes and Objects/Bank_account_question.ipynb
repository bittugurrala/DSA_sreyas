## Problem Statement

Create a **BankAccount** class with the following features:

### Data Members:

* Account Holder Name
* Account Number
* Balance

### Methods:

* Deposit Money
* Withdraw Money
* Check Balance

---

## Python Program

```python
class BankAccount:

    def __init__(self, name, account_number, balance):
        # Initialization of data members
        self.name = name
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        # Adds money to balance
        self.balance = self.balance + amount
        print(f"Amount Deposited: {amount}")
        print(f"Updated Balance: {self.balance}")

    def withdraw(self, amount):
        # Withdraw money if sufficient balance is available
        if amount <= self.balance:
            self.balance = self.balance - amount
            print(f"Amount Withdrawn: {amount}")
            print(f"Remaining Balance: {self.balance}")
        else:
            print("Insufficient Balance")

    def check_balance(self):
        # Displays current balance
        print(f"Account Holder: {self.name}")
        print(f"Account Number: {self.account_number}")
        print(f"Current Balance: {self.balance}")
```

---

## Creating an Object and Testing

```python
acc1 = BankAccount("Rahul", 12345, 5000)

acc1.check_balance()
acc1.deposit(2000)
acc1.withdraw(1000)
acc1.withdraw(10000)
```

---

## Output Explanation

* An account is created with an initial balance of 5000
* 2000 is deposited
* 1000 is withdrawn successfully
* Withdrawal of 10000 fails due to insufficient balance
