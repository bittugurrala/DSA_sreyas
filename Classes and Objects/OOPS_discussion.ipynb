```python
class Car:
    """
    Car class represents a real-world car with its properties and behaviors.
    
    Properties:
    - color: Color of the car
    - engine: Type of engine
    - mileage: Fuel efficiency of the car
    - model: Model name
    - manufactured_by: Manufacturer company
    - year: Manufacturing year

    Behaviors:
    - start()
    - stop()
    - accelerate()
    - brake()
    """

    #init method
    def __init__(self, color, engine, mileage, model, manufactured_by, year):
        self.color = color
        self.engine = engine
        self.mileage = mileage
        self.model = model
        self.manufactured_by = manufactured_by
        self.year = year
        self.speed = 0   # Tracks current speed of the car

    def start(self):
        """Starts the car"""
        print(f"The {self.model} car has started.")

    def stop(self):
        """Stops the car completely"""
        self.speed = 0
        print("The car has stopped.")

    def accelerate(self, increase_by):
        """Increases the speed of the car"""
        self.speed += increase_by
        print(f"The car is accelerated by {increase_by} km/h.")
        print(f"Current speed: {self.speed} km/h")

    def brake(self, decrease_by):
        """Decreases the speed of the car safely"""
        if self.speed > 0:
            self.speed -= decrease_by

            # Prevent negative speed
            if self.speed < 0:
                self.speed = 0

            print(f"The car is decelerated by {decrease_by} km/h.")
            print(f"Current speed: {self.speed} km/h")
        else:
            print("The car is already stopped. No need to apply brakes.")


# ===== Driver Code / Testing =====

ford = Car("Black", "V4", "45 km/l", "Mustang", "Ford", 1999)

print(f"Car Color: {ford.color}")
print(f"Car Model: {ford.model}")
print(f"Manufacturer: {ford.manufactured_by}")

ford.start()
ford.accelerate(10)
ford.brake(5)
ford.brake(5)
ford.brake(10)
ford.stop()
```

```python
class BankAccount:
    def __init__(self, name, number):
        self.name = name
        self.number = number
        self.pin = 8464
        self.balance = 0
        
    def create_pin(self, pin, password):
        self.pin = password
        print(f"Pin updated")
        
    def deposite(self, amount, password):
        if self.pin != password:
            print(f"Password incorrect!")
            return 
        self.balance += amount
        print(f"Amount {amount} is Credited!")
        return
    
    def withdraw(self, amount, password):
        if self.pin != password:
            print(f"Password incorrect!")
            return
        if self.balance < amount:
            print(f"Insufficient Funds")
            return
        elif self.balance <= 2000:
            print(f"Insufficient Funds")
            return 
        self.balance -= amount
        print(f"Amount {amount} debited")
        
    def check_balance(self, password):
        if self.pin != password:
            print(f"Password incorrect!")
            return
        print(f"""Name : {self.name}
balance : {self.balance}""")

acc1 = BankAccount("Sai", 123124)
acc1.create_pin(8464, "Sreyas@2025")
acc1.deposite(2000, "Sreyas@2025")

acc1.withdraw(1000, "Sreyas@2025")

acc1.check_balance("Sreyas@2025")



        
        
    
```
