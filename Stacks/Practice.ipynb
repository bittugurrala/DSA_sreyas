```python
def valid_parantheses(string):
    stack = []
    hashmap = {"}" : "{", "]" : "[", ")" :"("}
    for i in string:
        if i in hashmap and stack:
            top_ele = stack.pop()
            if top_ele != hashmap[i]:
                return False
                
        else:
            stack.append(i)
    return True
        

string = "[{{({})}}]"
print(valid_parantheses(string))
```

```python
def precedence(op):
    if op == '^':
        return 3
    if op in ('*', '/', '%'):
        return 2
    if op in ('+', '-'):
        return 1
    return 0

def infix_to_postfix(expr):
    stack = []
    postfix = ""

    for ch in expr:
        # 1. operand → add to postfix
        if ch.isalnum():
            postfix += ch

        # 2. '(' → push
        elif ch == '(':
            stack.append(ch)

        # 3. ')' → pop until '('
        elif ch == ')':
            while stack and stack[-1] != '(':
                postfix += stack.pop()
            stack.pop()  # remove '('

        # 4. operator
        else:
            while (stack and stack[-1] != '(' and
                   precedence(stack[-1]) >= precedence(ch)):
                postfix += stack.pop()
            stack.append(ch)

    # 5. pop remaining operators
    while stack:
        postfix += stack.pop()

    return postfix


# Example
exp = "A+B*C"
print(infix_to_postfix(exp))
```
